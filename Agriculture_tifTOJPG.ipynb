```python

!pip install rasterio matplotlib pillow

import rasterio
import numpy as np
from PIL import Image
from google.colab import files


uploaded = files.upload()
tif_file = list(uploaded.keys())[0]


with rasterio.open(tif_file) as src:
    band = src.read(1)


def apply_colormap(array):
    rgb = np.zeros((array.shape[0], array.shape[1], 3), dtype=np.uint8)
    rgb[array == 0] = [255, 0, 0]   # Red for class 0
    rgb[array == 1] = [0, 255, 0]   # Green for class 1
    return rgb

colored_image = apply_colormap(band)


img = Image.fromarray(colored_image)
jpg_file = tif_file.replace('.tif', '.jpg')
img.save(jpg_file)
print(f"Image saved as {jpg_file}")


files.download(jpg_file)
```
