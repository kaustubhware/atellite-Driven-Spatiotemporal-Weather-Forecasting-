```python
!pip install rasterio
!pip install numpy
!pip install matplotlib
import rasterio
import numpy as np
import matplotlib.pyplot as plt

# Load the NDVI image
file_path = '/content/Sentinel_NDVI_6_2024_to_10_2024.tif'

with rasterio.open(file_path) as src:
    ndvi_image = src.read(1)  # Read the first band
    ndvi_image = np.where(ndvi_image == src.nodata, np.nan, ndvi_image)  # Replace nodata with NaN

# Plot the image
plt.imshow(ndvi_image, cmap='RdYlGn')
plt.colorbar(label='NDVI')
plt.title('NDVI June to October 2024')
plt.show()
```
